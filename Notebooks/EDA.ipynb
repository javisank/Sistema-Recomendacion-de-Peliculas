## Data Wrangling: Getting columns for each of the movie genres.

```python
#movies['gen_lst'] = movies['genres'].apply(lambda x: x.split('|'))
lista = [sublista.split('|') for sublista in movies['genres']]
```

```python
genres = sorted(set([item for s in lista for item in s]))
genres
```

```python
dummies = pd.DataFrame(np.zeros((len(movies), len(genres)), dtype=int), columns=genres)
dummies.head()
```

```python
for i, gen in enumerate(movies['genres']):
    dummies.loc[i, gen.split('|')] = 1
```

```python
ratings['number_of_ratings'] = movie_rating.groupby('movieId')['rating'].count()
ratings.head()
```

```python
ratings = pd.merge(ratings,movies,on='movieId')
ratings.sample(10)
```
